# LangChain
### What is LangChain?
- framework - build apps using language models
- LM - AI tools - understand, generate, response to human language  

LangChain - intergrate LM into apps: chatbots, QA systems, summarization tools, rcm systems...

### LangChain's Core Capabilities
- Q-A Capabilities
- Summarization
- Keywords
- Multi-Agent Systems
