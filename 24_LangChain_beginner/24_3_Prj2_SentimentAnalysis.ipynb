### Understand Sentiment Analys in NLP
#### What is Sentiment Analysis?
- determine the emotional tone behind a piece of text
- classify text: positive/negative/ neutral
#### Applications
- industries: customer feedback, service on social media..
#### How SE works
- recognize patterns and asociate specific words and phrases --> certain sentiments


### SE - pretrained model in HF

```python
from transformers import  pipeline

# Load the SE pipeline
sentiment_pipeline = pipeline("sentiment-analysis", model = "distilbert-base-uncased-finetuned-sst-2-english")

def analyze_sentiment(text):
    response = sentiment_pipeline(text)
    sentiment = response[0]['label']
    confident = response[0]['score']
    return sentiment, confident

text = "I love working with langchain"
sentiment, confidence = analyze_sentiment(text)
print(f"Text: {text}")
print(f"Sentiment: {sentiment} (Confidence: {confidence:.2f})")

def main():
    print("Start SE bot")
    while (True):
        text = input("Enter text to analyze (or 'exit' to exit)")
        if text.lower() == 'exit':
            print("Exit")
            break
        sentiment, score = analyze_sentiment(text)
        print(f"Sentiment: {sentiment}(score: {score:.2f})")

main()
```
